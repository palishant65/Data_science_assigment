# International T20 Cricket Data Analysis
This project analyzes T20 cricket match data using Python libraries like Pandas, NumPy, Matplotlib and Seaborn.

# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Creating sample dataset

```python
data = {
 "team1":["India","Australia","England","Pakistan"],
 "team2":["England","India","Pakistan","India"],
 "winner":["India","Australia","England","Pakistan"],
 "run_margin":[20,15,10,5]
}

df = pd.DataFrame(data)
df
```

#Statistical Analysis

```python
df.describe()
```

```python
df["winner"].value_counts().plot(kind="bar")
plt.title("Most Successful Teams")
plt.show()
```
